```python
import polars as pl
```

```python
def merge_cpi_m2(cpi_csv, m2_csv):
    # Read the CSV files
    cpi_df = pl.read_csv(cpi_csv)
    m2_df = pl.read_csv(m2_csv)

    # Lag the Money Supply data by 2 years (24 months)
    m2_df = m2_df.with_columns(
        pl.col('value').shift(24).alias('m2_lagged')
    )

    # Merge the two DataFrames on the 'date' column
    merged_df = cpi_df.join(m2_df.select(['date', 'm2_lagged']), on='date', how='outer')

    # Save the merged DataFrame to a new CSV file
    merged_df.write_csv('output/merged_cpi_m2.csv')
    print("Merged CPI and M2 Money Supply data saved to 'output/merged_cpi_m2.csv'")
```

```python
import os

# Set your desired working directory
cpi_csv_f = "/Users/bradleystrong/Desktop/REPOS/AetherLoom/AetherLoom/output/GS30_data.csv"
m2_csv_f = "/Users/bradleystrong/Desktop/REPOS/AetherLoom/AetherLoom/output/G2_data.csv"

merge_cpi_m2(cpi_csv_f, m2_csv_f)
```
